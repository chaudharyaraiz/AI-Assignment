```python
# assignment1_eda.py
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class LaptopPriceEDA:
    def __init__(self, csv_path):
        self.df = pd.read_csv(csv_path)
        print("‚úÖ Data Loaded Successfully")

    def perform_eda(self):
        print("\nüìä Dataset Info:")
        print(self.df.info())
        print("\nüîç Missing Values:")
        print(self.df.isnull().sum())

        # Company distribution
        plt.figure(figsize=(10, 5))
        sns.countplot(data=self.df, x='Company')
        plt.title("Company Distribution")
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()

        # Price distribution
        plt.figure(figsize=(8, 5))
        sns.histplot(self.df['Price'], kde=True)
        plt.title("Price Distribution")
        plt.tight_layout()
        plt.show()

        # RAM vs Price boxplot (ensure Ram column exists and is numeric or string '8GB' style)
        if 'Ram' in self.df.columns:
            plt.figure(figsize=(8, 5))
            sns.boxplot(x='Ram', y='Price', data=self.df)
            plt.title("RAM vs Price")
            plt.tight_layout()
            plt.show()

    def save_preview(self, out_csv='preview_laptop_data.csv', n=200):
        """
        Save a small preview of the dataset (useful for handing to Assignment 2).
        """
        self.df.head(n).to_csv(out_csv, index=False)
        print(f"‚úÖ Preview saved to {out_csv}")

# Example usage:
if __name__ == "__main__":
    eda = LaptopPriceEDA('laptop_data.csv')  # replace with your CSV path
    eda.perform_eda()
    eda.save_preview('preview_laptop_data.csv')
```
